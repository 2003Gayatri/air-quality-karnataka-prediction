Strong correlations (for example, correlations above 0.6 or below -0.6), we can do two things:

Mask weak correlations

Use bold colors for strong ones

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Select pollutant columns
pollutants = ['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)',
              'SO2 (µg/m³)', 'CO (mg/m³)', 'Ozone (µg/m³)']

# Compute correlation matrix
corr = df[pollutants].corr()

# Mask weak correlations: keep only values with abs(corr) > 0.6
strong_corr = corr.where((corr.abs() >= 0.6))

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    strong_corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    annot_kws={"size": 10},
    vmin=-1, vmax=1
)

plt.title("Heatmap: Strong Correlations (|corr| ≥ 0.6)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

HISTOGRAM FOR NUMERICAL VALUES 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (ensure the CSV is in the same folder or give full path)
df = pd.read_csv('D:\\Internship\\Air-Quality-Analysis\\Github-air data\\air-quality-karnataka-prediction\\Air_data\\2024.csv')

# --- BASIC INFO ---
print("=== Basic Info ===")
print(df.info())
print("\n=== Missing Values ===")
print(df.isnull().sum())

# Set seaborn style
sns.set(style='whitegrid')

# --- HISTOGRAMS FOR NUMERIC COLUMNS ---
num_cols = df.select_dtypes(include='number').columns
df[num_cols].hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Columns", fontsize=16)
plt.tight_layout()
plt.show()





```

BOX PLOT FOR NUMERICAL VALUES 

```python
# --- BOXPLOTS FOR NUMERIC COLUMNS (SAFE CHECKS) ---
# Keep only columns with at least one non-NaN value
num_cols = [col for col in num_cols if df[col].dropna().shape[0] > 0]

for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```
